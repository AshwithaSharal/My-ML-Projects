```python
classes = []
label_number=0
audio_data = []
labels = []
sampling_rate = []
file_names = []
data = []
noisy_removed=[]
noise=[]


for j in final_list:
    y, s = librosa.load(j, sr=44000) # Downsample 44.1kHz to 8kHz
    data.append(y)
```

```python
from sklearn.decomposition import PCA

pca_2 = PCA(n_components=100)

wavelet=[]
for x in range(0,len(data)):

    scales = np.arange(1, 101)
    coeff1, freqs1 = pywt.cwt(data[x][0:60000], scales, 'morl')
    # print(x)
    # print(coeff1.shape)

    # print(pca_2.fit_transform(coeff1).shape)
    wavelet.append(pca_2.fit_transform(coeff1))
    #wavelet.append((coeff1))

wavelet=np.array(wavelet)
print(wavelet.shape)
```

```python
wavelet.shape
```

```python
wavelet_data=wavelet.reshape(2448,-1)
# wavelet_data=wavelet
wavelet_data.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))
scaler.fit(wavelet_data)


print(np.amax(wavelet_data))

normalized_features = scaler.transform(wavelet_data)
print(np.amax(normalized_features))
```

```python
label_list=np.array(label_list)
```

```python
label_list.shape
```

```python
normalized_features.shape
```

```python
from sklearn import tree
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
X_train, X_test, y_train, y_test = train_test_split(normalized_features, label_list, test_size=0.5, random_state=1150)
```

```python
clf = tree.DecisionTreeClassifier()
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
print(accuracy_score(y_test, y_pred))
report = classification_report(y_test, y_pred)
print(report)
```
